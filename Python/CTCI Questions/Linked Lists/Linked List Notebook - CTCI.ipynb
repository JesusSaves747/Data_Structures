#### 2.2 Find the kth from last node: 

```python
## Function that returns the data in the kth from last node: 

# End cases: 1. Length of list is less than k: 

# Pre-condition: Input given is a Linked List object and the value of k

def kth_from_last(LL, k): 
    
    # Initialize two cursors and a counter: 
    fast = LL.head
    slow = LL.head 
    count = 0
    
    while fast: 
        
        # Move the slow cursor forward only once the fast one has moved k nodes forward:
        if count >= k:
            slow = slow.next        
        
        # Move the fast cursor forward unitl it reaches the end: 
        fast = fast.next       
        
        # Incerment the counter: 
        count +=1
        
    return slow.data
    

### Test out the implementation: 

# Create a sample linked list: 
import numpy as np

LL = LinkedList(0)

numList = np.arange(1,10)

np.random.shuffle(numList)

cursor = LL.head

for num in numList: 
        
        cursor.next = Node(num)
        
        cursor = cursor.next
        

print (" The Linked List is : ")

LL.print()

k = int(input("Enter the value of k"))

res = kth_from_last(LL, k)

print (" The " , k, "th last element is: " , res)

    
```

```python
### Strategy: Given access only to the middle node what we have to do is shift the data from all the forward nodes one step back
#             and then delete the last node.
def del_mid(midNode): 
    
    # Initialize a cursor as the mid Node
    cursor = midNode  

    
    # Shift data from next node to the current node:
    cursor.data = cursor.next.data
        
    # Now delete the node ahead of the current node:
    temp = cursor.next
    
    cursor.next = cursor.next.next
    
    temp = None
    

## Test out the implementation:
midNode = LL.get_kth_node(5)

LL.print()

del_mid(midNode)

print (" After deletion: ")

LL.print()

## Comments on: 
# 1. Time Complexity: 

# 2. Space Complexity:
```

#### 2.8 Loop detection in a Linked List: 

```python
### Given a Circular linked list. Find the node at which the loop begins: 

# Use two pointers: One moving one node at a time and the other 2 nodes at a time:
# Assume the loop start is k nodes into the list. Thus, when slow is a loop start fast is at p= mod(k,LOOP_size) nodes ahead of loop start
# i.e  fast is (LOOP_SIZE - p) nodes away from loop start. 
# Move both pointers at their respective speeds until they intersect. When will this happen? 
# Since fast is catching up with slow at the rate of 1 node by step of slow. The intersection will happen when slow has moved
# (LOOP_SIZE - p) steps. Thus, both slow and fast are now 'p' steps away from loop start.But even the head node of the list is 
# c*LOOP_SIZE + p steps away from the loop start. 
# Move one of the pointers to the head node. And now move each forward by one step at a time until intersection. This is the loop start: 

# Pre-condition: Input is a linked list that may or may not have a loop:
# Post-condition: If the list has a loop then return the node at which the loop begins else return None
def loop_detect(LL): 
    
    # Initialize the slow and fast pointers: 
    slow = LL.head
    fast = LL.head
    
    # Variable to facilitate different speeds: 
    moveSlow = False 
    
    # Variable to detect the loop: 
    loopExists = False
    
    
    # Keep going until the end of the list:
    while fast and slow: 
        
        # Conditionally move the slow pointer ahead
        if moveSlow:             
            slow = slow.next 
            moveSlow = False
            
        else:             
            moveSlow = True
        
        # Always move the fast pointer ahead: 
        fast = fast.next
        
        
        
        # Check to see if the two pointers have intersected: 
        if fast == slow:            
            loopExists = True
            break
        

    
    # Coming out of the loop see if Loop exists or not: 
    if loopExists:
        
        newp = LL.head
        
        while True:        
            
            if fast == newp:
                break
            
            fast = fast.next 
            newp = newp.next
            
        
        return fast
        
    # If there is no loop: 
    else: 
        
        return None

    
### Test this out: How to generate a circular linked list:
LL = generate_random_list(15,26)

LL.print()
print(" ")

loopStart = LL.get_kth_node(4)

cursor = LL.head

while cursor.next: 
    
    cursor = cursor.next
    

# Make it circular: 
cursor.next = loopStart

res = loop_detect(LL)

print (res.data)

if res is not None: 
    
    print (" Loop exists at : ", res.data)

else: 
    
    print (" Loop DOES NOT exist")
```
