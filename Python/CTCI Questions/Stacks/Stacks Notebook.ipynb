```python
## Import all modules: 
import numpy as np
```

```python
### Class for nodes in a stack: 
class Snode:
    
    # Constructor: 
    def __init__(self, data):
        
        # Three parameters: 
        # 1. data - Data in the stack: 
        # 2. next - Pointer to the node below 
        # 3. min - Current minimum element in all the nodes below: 
        self.data = data
        self.next = None
        self.min = None
    
### Class for Stacks: Implementing the Stack as a Linked List: 
class Stack:
    
    # Constructor: 
    def __init__(self):
        
        # Single parameter: Pointer to the top node: 
        self.top = None
        
        # Variable to indicate current size: 
        self.size = 0
        
    # Function to push a new element onto the stack: 
    def push(self, new): 
        
        # Check if the Stack is empty: 
        if self.top is None: 
            
            self.top = Snode(new)
            
            self.top.min = new
            
        else: 
            
            temp = Snode(new)
            
            temp.next = self.top
            
            # Update the current minimum element in the 
            if new < self.top.min: 
                
                temp.min = new
            
            else: 
                
                temp.min = self.top.min
            
            
            self.top = temp
            
        self.size +=1
            
            
            
    # Function to pop the top element from the stack: 
    def pop(self): 
        
        # Extract the data from the top node: 
        dataPop = self.top.data
        
        # Remove the top most node: 
        temp = self.top 
        
        # Move the pointer one step below: 
        self.top = self.top.next
        
        # Delete the node: 
        temp = None
        
        # Decrement the size: 
        self.size -=1
        
        
        return dataPop
    
    # Function to return the minimum element in the stack: 
    def get_min(self): 
        
        return self.top.min
    
    # Function to print all the elements in the stack: Starting from the top element: 
    def print(self): 
        
        # Initialize a cursor: 
        cursor = self.top
        
        while cursor: 
            
            print(cursor.data ,end = " ->" )
            
            cursor = cursor.next   
    
    
## Define a utility function that generates a Stack with random integers: 
def generate_random_stack(start, stop, step):
    
    SS = Stack()
    
    numList = np.arange(start, stop, step)

    np.random.shuffle(numList)

    for num in numList: 

        SS.push(num)
        
    return SS        


SS = generate_random_stack(12,54,3)

print (" Behold thy Stack!!!")
SS.print()


# Remove a few elements now: 
SS.pop()

SS.pop()


min_elem = SS.get_min()

print (" Current minimum element is:" , min_elem)
SS.print()
        
        
```

```python

class SetOfStacks:
    
    
    # Constructor: 
    def __init__(self):
        
        ### Parameters: 
        # 1. Capcity of each stack: 
        self.CAPACITY = 10

        # Create a list of stacks: 
        self.Set = []

        # Indicator variable to indicate whic stack we are on: 
        self.stackIdx = None

    # Push function of the set:
    def push(self,new): 
        
        # Check if the Set of Stacks is empty: 
        if not self.Set: 
            
            # Create the first stack: 
            SS = Stack()
            
            SS.push(new)

            self.Set.append(SS)
            
            self.stackIdx =0
            
            print (" Created first stack!")
            
        else: 
            
            if self.Set[self.stackIdx].size < self.CAPACITY: 
                
                self.Set[self.stackIdx].push(new)
                
            else: 
                
                # Create a new Stack: 
                SS = Stack()
                
                # Push the new element on the new stack: 
                SS.push(new)
                
                # Append the new Stack to the set: 
                self.Set.append(SS)
                
                # Increment the Stack index: 
                self.stackIdx +=1

                
    # Function to pop an element from the set of Stacks: 
    def pop(self): 

        # Remove the top element from the current stack: 
        dataPop = self.Set[self.stackIdx].pop()

        # Check if the size of the current stack has gone to zero: 
        if self.Set[self.stackIdx].size == 0: 

            # Remove the stack from the set: 
            self.Set.pop(self.stackIdx)

            # Decrement the stack index: 
            self.stackIdx -= 1


        return dataPop
    
    # Function to print the elements in the Set of stacks: 
    def print(self): 
        
        # Loop through the different stacks in the set: 
        for i in range(0, self.stackIdx + 1):
            
            self.Set[i].print()
            
            print (" -- ")
            

        
### Test this out: 
mySet = SetOfStacks()

numList = np.arange(1,32)


for num in numList: 
    
    mySet.push(num)


    # Original set is: 
print (" Original Set is: ")
mySet.print()

mySet.pop()

mySet.pop()

mySet.pop()

# Original set is: 
print ("  After popping : ")
mySet.print()

mySet.push(11)

mySet.push(12)

mySet.push(13)

mySet.push(14)

print ("  After pushing again : ")
mySet.print()
                
                
```
